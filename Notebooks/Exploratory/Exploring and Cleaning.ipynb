```python
df3['production_budget'] = df3['production_budget'].str.replace(',', '')
df3['production_budget'] = df3['production_budget'].str.replace('$', '')
df3['production_budget'] = df3['production_budget'].astype(int)
```

```python
df3['domestic_gross'] = df3['domestic_gross'].str.replace(',', '')
df3['domestic_gross'] = df3['domestic_gross'].str.replace('$', '')
df3['domestic_gross'] = df3['domestic_gross'].astype(int)
```

```python
df3['worldwide_gross'] = df3['worldwide_gross'].str.replace(',', '')
df3['worldwide_gross'] = df3['worldwide_gross'].str.replace('$', '')
df3['worldwide_gross'] = df3['worldwide_gross'].astype('int64')
```
